## Sushi Pairs on Avalanche and BSC

 1. Create a document that includes all addresses for all Swap pairs on **Avalanche** and **BSC** for sushi, plus the addresses of the two tokens belonging to each pair.

### 1.1 Swap Pairs on BSC for Sushi 

```python
#DataFrame to Store the data
fields = ['POOL_NAME', 'SYMBOL_0', 'SYMBOL_1','POOL_ADDRESS', 'TOKEN_0', 'TOKEN_1']
df = pd.DataFrame(columns =  fields )

#Web Scrapping Setting
print('Starting Web Scrapping')
options = Options()
options.add_argument("window-size=800,600")
options.add_argument("disable-infobars")
options.add_argument("--disable-extensions")
driver = webdriver.Chrome(options=options, executable_path='chromedriver')
driver.get("https://app.sushi.com/analytics/pools?chainId=56")

#Looping through next pages
for page in range(1,172):
    print('Page {}'.format(page))
    #Looping through items  in the same page
    for LP in range(1,11):
        try:
            WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, """//*[@id="__next"]/div/main/main/div/div/div[2]/div[2]/table/tbody/tr[{}]""".format(LP))))
            driver.find_element(By.XPATH,"""//*[@id="__next"]/div/main/main/div/div/div[2]/div[2]/table/tbody/tr[{}]""".format(LP)).click()

            time.sleep(2)
            #Collect LP and Token Name
            names = []
            for i in range(1,4):
                WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH,"/html/body/div[1]/div/main/main/div/div/div[7]/div/table/thead/tr/th[{}]".format(i))))
                name = driver.find_element(By.XPATH,"/html/body/div[1]/div/main/main/div/div/div[7]/div/table/thead/tr/th[{}]".format(i)).text
                names.append(name.split('Address')[0].strip())

            #Collect LP and Token addresses
            addresses = []
            for j in range(1,4):
                WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, "/html/body/div/div/main/main/div/div/div[7]/div/table/tbody/tr/td[{}]".format(j))))
                address = driver.find_element(By.XPATH,"/html/body/div/div/main/main/div/div/div[7]/div/table/tbody/tr/td[{}]".format(j)).text
                addresses.append(address.strip())
            #Append data to the dataset
            row = names + addresses
            df =df.append(dict(zip(fields,row)), ignore_index = True)
            print(names[0] + ' added to the dataframe')

            #Move backward in the browser
            driver.back()
            time.sleep(3)

        except Exception as e:
            print(e)
            #Move backward in the browser
            driver.back()
            time.sleep(6)
    #Next page
    #driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(0.5)
    WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, "/html/body/div[1]/div/main/main/div/div/div[2]/div[4]/div[2]/div[2]")))
    driver.find_element(By.XPATH,  "/html/body/div[1]/div/main/main/div/div/div[2]/div[4]/div[2]/div[2]").click()
    time.sleep(3)
    


#driver.close()
print('Web Scrapping Process ended')
```
