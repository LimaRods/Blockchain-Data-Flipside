## What makes a Top Shots moment valuable? (Part I)

### 1. Loading and Preparing Data

```python
API_KEY = "API_KEY"

SQL_QUERY = """
WITH NBA_SALES AS (
SELECT
	DATE_TRUNC('day', SALES.BLOCK_TIMESTAMP) AS DATE,
  	NBA. PLAY_TYPE,
	SALES.CURRENCY,
  	SUM(SALES.PRICE) AS  DAILY_VOL

FROM
	flow.core.fact_nft_sales SALES
	INNER JOIN flow.core.dim_topshot_metadata NBA ON SALES.NFT_ID  = NBA.NFT_ID

WHERE TX_SUCCEEDED = TRUE
  --AND NFT_COLLECTION IN (SELECT NFT_COLLECTION FROM flow.core.dim_topshot_metadata)
GROUP BY 1,2,3
ORDER BY 1,2,3),
  
TOKEN_PRICES AS (
	SELECT
  		DATE_TRUNC('day',TIMESTAMP) AS DATE,
  		TOKEN,
  		TOKEN_CONTRACT,
  		AVG(PRICE_USD) AS DAILY_PRICE
  	FROM
  		flow.core.fact_prices
  	WHERE TOKEN_CONTRACT IN (SELECT CURRENCY FROM NBA_SALES)
  	GROUP BY  1,2,3
  	ORDER BY 1		
),

NBA_SALES_USD AS (
(SELECT
  NBA.DATE,
  PLAY_TYPE,
  CURRENCY,
  DAILY_VOL,
  (CASE
  	WHEN TOKEN  IS NULL THEN 'DAPP'
  	ELSE TOKEN
  END) AS TOKEN,
  (CASE
  	WHEN DAILY_PRICE IS NULL THEN 1
  	ELSE DAILY_PRICE
  END) AS DAILY_PRICE
  
FROM
  	NBA_SALES NBA
	LEFT OUTER JOIN TOKEN_PRICES P ON  NBA.CURRENCY = P.TOKEN_CONTRACT 
	WHERE CURRENCY NOT IN (SELECT TOKEN_CONTRACT FROM TOKEN_PRICES) -- Explain DapperUtilityCoin is equal to 1 usd
ORDER BY 1)
UNION ALL
  (SELECT
    NBA.DATE,
    PLAY_TYPE,
    CURRENCY,
    DAILY_VOL,
   TOKEN,
  DAILY_PRICE 
  FROM
    	NBA_SALES NBA
  	INNER JOIN TOKEN_PRICES P ON  NBA.CURRENCY = P.TOKEN_CONTRACT 
  	WHERE NBA.DATE = P.DATE
  ORDER BY 1))

SELECT
	DATE,
    PLAY_TYPE,
  	sum(DAILY_VOL * DAILY_PRICE) AS  DAILY_VOL_USD
FROM
	NBA_SALES_USD
group by 1,2
ORDER BY 1
"""

TTL_MINUTES = 15

# return up to 100,000 results per GET request on the query id
PAGE_SIZE = 100000

# return results of page 1
PAGE_NUMBER = 1

def create_query():
    r = requests.post(
        'https://node-api.flipsidecrypto.com/queries', 
        data=json.dumps({
            "sql": SQL_QUERY,
            "ttlMinutes": TTL_MINUTES
        }),
        headers={"Accept": "application/json", "Content-Type": "application/json", "x-api-key":API_KEY},
    )
    if r.status_code != 200:
        raise Exception("Error creating query, got response: " + r.text + "with status code: " + str(r.status_code))
    
    return json.loads(r.text)    


def get_query_results(token):
    r = requests.get(
        'https://node-api.flipsidecrypto.com/queries/{token}?pageNumber={page_number}&pageSize={page_size}'.format(
          token=token,
          page_number=PAGE_NUMBER,
          page_size=PAGE_SIZE
        ),
        headers={"Accept": "application/json", "Content-Type": "application/json", "x-api-key": API_KEY}
    )
    if r.status_code != 200:
        raise Exception("Error getting query results, got response: " + r.text + "with status code: " + str(r.status_code))
    
    data = json.loads(r.text)
    if data['status'] == 'running':
        time.sleep(10)
        return get_query_results(token)

    return data


def run():
    query = create_query()
    token = query.get('token')
    data = get_query_results(token)

    #print(data['columnLabels'])
    list_data = []
    for row in data['results']:
        #print(row)
        list_data.append(row)
    df = pd.DataFrame(data =list_data, columns = data['columnLabels'])
    return df
df_tx = run()
```

### 2. Applying One-Way ANOVA

```python
df_tx.head()
```

```python
#Test of normality
#Using Shapiro-Wilk test to evaluate my data population in order to figure out whether or not my sample data proceeds
#a Gaussian distribution
stat,p = stats.shapiro(df_tx["DAILY_VOL_USD"]) 
alpha = 0.05
if  p < alpha:
    print("p-value {} and test statistic {}. The data dont'n follow a Gaussian distribution".format(p, stat))
if p > alpha:
    print("p-value {} and test statistic {}. I can't conclude the data do not follow a Gaussian shape".format(p,stat))
```

```python
#Descriptive Statistics
def describe_and_corr(df,field,variable):
    "df: dataset"
    "field: what categorical variable"
    ""
    stat = round(df.groupby(field).agg({variable: ['count','max','min','mean','std','var']}),4)
    scores = df[variable].to_numpy()
    stat = stat.append(pd.DataFrame([[np.count_nonzero(scores),np.max(scores),np.min(scores),np.mean(scores), np.std(scores), np.var(scores)]], index = ['Total'], columns = stat.columns))
    
    #Computting the correlation between variables
    stat['weighted var'] = stat[(variable,'count')] * stat[(variable,'var')]
    #stat.iloc[-1,-1] == np.nan
    weighted_var = np.sum(stat.iloc[0:-1,-1])/(stat.iloc[-1,0] - 4)
    R2 = 1 - (weighted_var/stat.iloc[-1,-2])
    return (stat,"R2:{}".format(R2))
R2_play_type = describe_and_corr(df_tx,'PLAY_TYPE',"DAILY_VOL_USD")
R2_play_type[0]
```

```python
#Reindexing the dataframe
def sample(n,field,group,df):
    'n: sample size'
    'field: variable'
    'groups: level of the nominal variable'
    data = df[['DATE','DAILY_VOL_USD',field]]
    sample = df.copy().sort_values(by = field, ascending = True).set_index([field,"DATE"])
    #display(sample)

    #Setting up samples to apply One-way ANOVA
    dataset = {}
    for i in group:
        dataset.setdefault(i,sample.loc[i].sample(n = n, replace = True)['DAILY_VOL_USD'].values)
    #display(dataset)
    samples = pd.DataFrame(dataset)
    return samples
play_type_sample = sample(41,'PLAY_TYPE',play_type_list,df_tx)
```

```python
#One-Way ANOVA
F,p = stats.f_oneway(play_type_sample.iloc[:,0],play_type_sample.iloc[:,1],play_type_sample.iloc[:,2],play_type_sample.iloc[:,3],play_type_sample.iloc[:,4],
                     play_type_sample.iloc[:,5], play_type_sample.iloc[:,6], play_type_sample.iloc[:,7], play_type_sample.iloc[:,8])
alpha = 0.05
if p < alpha:
    print("""
    F = {}, p-value = {}.
    I reject null hypothesis Ho.
    At least one popullation mean is different from others""".format(F,p))
if p > alpha:
     print("""
     F = {}, p-value = {}.
     I don't have enough evidence to reject  null hypothesisHo.
     All population means are statistica""".format(F,p))
```
